```python
import pandas as pd
import numpy as np
from os.path import join
import os
from scipy.stats import gmean
import matplotlib.pyplot as plt
import matplotlib as mpl

plt.style.use('CCB_plot_style_0v4.mplstyle');
c_styles      = mpl.rcParams['axes.prop_cycle'].by_key()['color']   # fetch the defined color styles
high_contrast = ['#004488', '#DDAA33', '#BB5566', '#000000']


import warnings
warnings.filterwarnings('ignore')

CURRENT_DIR = os.getcwd()

df_kcat = pd.read_pickle(join("..", "..", "data", "kcat_data", "merged_and_grouped_kcat_dataset_with_FPs_and_ESM1bs.pkl"))
UNIPROT_df = pd.read_csv(join("..", "..", "data", "enzyme_data",  "uniprot_results_eukaryotes_versus_prokaryotes.tsv"),
                         sep = "\t")
UNIPROT_df.drop(columns = ["Entry"], inplace = True)
display(UNIPROT_df.head())
```

```python
from ete3 import NCBITaxa
from Bio import Entrez
ncbi = NCBITaxa()
#ncbi.update_taxonomy_database()
```

```python
def get_lineage(tax_id):
    handle = Entrez.efetch(db="Taxonomy", id=tax_id, retmode="xml")
    records = Entrez.read(handle)
    lineage = records[0]["Lineage"].split("; ")
    return(lineage)

def get_Division(tax_id):
    handle = Entrez.efetch(db="Taxonomy", id=tax_id, retmode="xml")
    records = Entrez.read(handle)
    return(records[0]["Division"])
```

```python
UNIPROT_df["Organism type"] = np.nan
for ind in UNIPROT_df.index:
    if ind >=-1:
        print(ind)
        lineage = get_lineage(tax_id = UNIPROT_df["Organism (ID)"][ind])
        if "Eukaryota" in lineage:
            UNIPROT_df["Organism type"][ind] = "Eukaryota"
        elif "Bacteria" in lineage or "Archaea" in lineage:
            UNIPROT_df["Organism type"][ind] = "Prokaryota"
```

```python
UNIPROT_df["Divison"] = ""

for ind in UNIPROT_df.index:
    UNIPROT_df["Divison"][ind] = get_Division(tax_id = UNIPROT_df["Organism (ID)"][ind])
    print(ind, UNIPROT_df["Divison"][ind])
```

```python
UNIPROT_df = pd.read_csv(join("..", "..", "data", "enzyme_data",  "uniprot_results_eukaryotes_versus_prokaryotes_with_type.tsv"),
                         sep = "\t")
```

```python
df = pd.DataFrame(columns = ["Reaction IDs", "kcats", "Uniprot IDs", "Organisms", "Oragnism type", "count"])

str_fps = list(set(df_kcat["structural_fp"]))

added_fps = []
for fp in str_fps:
    help_df = df_kcat.loc[df_kcat["structural_fp"] == fp]
    
    if len(help_df) > 1:
        organisms = []
        UIDs = []
        kcats = []
        types = []
        
        try:
            for ind in help_df.index:

                seq = help_df["Sequence"][ind]
                UIDs.append(list(UNIPROT_df["From"].loc[UNIPROT_df["Sequence"] == seq])[0])
                organisms.append(list(UNIPROT_df["Organism"].loc[UNIPROT_df["Sequence"] == seq])[0])
                types.append(list(UNIPROT_df["Organism type"].loc[UNIPROT_df["Sequence"] == seq])[0])
                kcat_values = [float(kcat) for kcat in help_df["kcat_values"][ind]]
                kcats.append(kcat_values)

            df = df.append({"Reaction IDs" : list(help_df["Reaction ID"]),
                            "kcats": kcats,
                            "Uniprot IDs" : UIDs,
                            "Organisms": organisms,
                            "Oragnism type" : types,
                            "count" : len(help_df)}, ignore_index = True)
        except IndexError:
            pass
df = df.sort_values(by = "count", ascending = False)
df.reset_index(drop = True, inplace = True)
df
```

```python
df["count_eu"] = 0
df["count_pro"] = 0
df["kcats_eu"], df["kcats_pro"] = "", ""
df["kcat_mean_eu"], df["kcat_mean_pro"] = "", ""
df["kcat_var_eu"], df["kcat_var_pro"] = "", ""

for ind in df.index:
    types, kcats = df["Oragnism type"][ind], df["kcats"][ind]
    select_eu = np.array(types) == "Eukaryota"
    select_pro = np.array(types) == "Prokaryota"
    kcats_eu, kcats_pro = np.array(kcats)[select_eu], np.array(kcats)[select_pro]
    
    if len(kcats_eu) > 0:
        df["count_eu"][ind] = len(kcats_eu)
        
        
        kcat_list_eu = []
        for kcats in kcats_eu:
            kcat_list_eu = kcat_list_eu + list(kcats)
            
        df["kcats_eu"][ind] = kcat_list_eu
        df["kcat_mean_eu"][ind] = np.mean(kcat_list_eu)
        df["kcat_var_eu"][ind] = np.std(kcat_list_eu)
        
    if len(kcats_pro) > 0:
        df["count_pro"][ind] = len(kcats_pro)
        
        
        kcat_list_pro = []
        for kcats in kcats_pro:
            kcat_list_pro = kcat_list_pro + list(kcats)
        
        df["kcats_pro"][ind] = kcat_list_pro
        df["kcat_mean_pro"][ind] = np.mean(kcat_list_pro)
        df["kcat_var_pro"][ind] = np.std(kcat_list_pro)
```

```python
fig, ax = plt.subplots(figsize= (8,8))
plt.rcParams.update({'font.size': 28})

df2 = df.loc[df["count_eu"] >= 2].loc[df["count_pro"] >= 2]

kcat_pro = [np.log10(kcat) for kcat in np.array(df2["kcat_mean_pro"])]
kcat_eu =  [np.log10(kcat) for kcat in np.array(df2["kcat_mean_eu"])]
ax.tick_params(axis='x', length=10)
ax.tick_params(axis='y', length=10)

ax.yaxis.set_label_coords(-0.13, 0.5)
ax.xaxis.set_label_coords(0.5, -0.1)

#plt.scatter(kcat_pro, kcat_eu)
plt.scatter(kcat_eu, kcat_pro, alpha = 0.7, s=80)
plt.plot([-3,6], [-3,6], c = "red")
```

```python
from scipy.stats import mannwhitneyu

res = mannwhitneyu(kcat_eu, kcat_pro, alternative="less")
res, np.mean(np.array(kcat_pro) > np.array(kcat_eu))
```
