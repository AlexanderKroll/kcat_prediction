```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn import metrics
from scipy import stats
from scipy.stats import wilcoxon
from sklearn.metrics import roc_auc_score, r2_score
from sklearn.linear_model import LinearRegression
import scipy
import os
from os.path import join
import pandas as pd

CURRENT_DIR = os.getcwd()
print(CURRENT_DIR)

import warnings
warnings.filterwarnings("ignore")

plt.style.use('CCB_plot_style_0v4.mplstyle');
c_styles      = mpl.rcParams['axes.prop_cycle'].by_key()['color']   # fetch the defined color styles
high_contrast = ['#004488', '#DDAA33', '#BB5566', '#000000']
```

```python
#TurNuP data
model = 'ESM1b_ts_DRFP_mean'
data_test = pd.read_pickle(join("..", "..", "data", "kcat_data", "splits", "test_df_kcat.pkl"))
pred_y = np.load(join("..", "..", "data", "training_results", "y_test_pred_xgboost_" + model + ".npy"))
test_y = np.load(join("..", "..", "data", "training_results", "y_test_true_xgboost_" + model + ".npy"))
data_test["y_true"] = np.round(test_y,5)
data_test["y_pred"] = np.round(pred_y,5)

#ENKIE data
df =pd.read_csv("C:\\Users\\alexk\\Downloads\\kcat_4.csv")
```

```python
fig, ax = plt.subplots(figsize= (8,8))
plt.rcParams.update({'font.size': 28})

test_y, pred_y = np.array(df["error"]), np.array(df["pred_error"])


plt.ylim(ymax = 4, ymin = -0.2)
plt.xlim(xmax = 4, xmin = -0.2)

ax.tick_params(axis='x', length=10)
ax.tick_params(axis='y', length=10)
ax.text(0.3, 3, r'$R^2=-0.05$', fontsize=26, c = "black")


ax.yaxis.set_label_coords(-0.18, 0.5)
ax.xaxis.set_label_coords(0.5, -0.1)

plt.xticks([0,1,2,3], [str(i) for i in range(4)])
plt.yticks([0,1,2,3], [str(i) for i in range(4)])



plt.ylabel("Predicted absolute error for \n $k_{cat}$ prediction on $\log_{10}$-scale", fontsize = 22)
plt.xlabel("Actual absolute error for \n$k_{cat}$ prediction on $\log_{10}$-scale", fontsize = 22)
plt.scatter(test_y, pred_y, alpha = 0.6, s=30, c="darkblue")
plt.savefig(join("C:\\Users\\alexk\\Downloads", "scatter_plot_errors_ENKIE.png"))
plt.show()
```

```python
fig, ax = plt.subplots(figsize= (8,8))
plt.rcParams.update({'font.size': 28})

test_y, pred_y = np.array(df["value"]), np.array(df["pred_value"])


plt.ylim(ymax = 5.1, ymin = -3.5)
plt.xlim(xmax = 5.1, xmin = -3.5)

ax.tick_params(axis='x', length=10)
ax.tick_params(axis='y', length=10)

ax.yaxis.set_label_coords(-0.18, 0.5)
ax.xaxis.set_label_coords(0.5, -0.1)

plt.xticks([-2,0,2,4], ["$10^{-2}$", "$10^{0}$", "$10^{2}$", "$10^{4}$"])
plt.yticks([-2,0,2,4], ["$10^{-2}$", "$10^{0}$", "$10^{2}$", "$10^{4}$"])



reg = LinearRegression().fit(test_y.reshape(-1,1), pred_y.reshape(-1,1),)
reg.score(test_y.reshape(-1,1), pred_y.reshape(-1,1))
beta0, beta1 =reg.intercept_[0], reg.coef_[0][0]
plt.plot([-3.5,4.9], [-3.5,4.9], color='grey', alpha = 0.3, linestyle='dashed')
plt.plot([-3.5,5.1], [beta0 + -3.5*beta1, beta0 + 5.1*beta1], color='red', alpha = 0.6, linestyle='dashed')

plt.ylabel("Predicted $k_{cat}$-values [$s^{-1}$]", fontsize = 22)
plt.xlabel("Empirical mean of measured $k_{cat}$-values [$s^{-1}$] \n \
for same enzyme-reaction pairs", fontsize = 22)
plt.scatter(test_y, pred_y, alpha = 0.6, s=30, c="darkblue")
plt.savefig(join("..","..", "data", "figures", "scatter_plot.eps"))
plt.show()
```

```python
r2_score(df["error"], df["pred_error"]), r2_score(df["value"], df["pred_value"]), np.mean(abs(df["value"]- df["pred_value"])**2)
```

```python
data_test["y_pred_ENKIE"]= np.nan
for ind in df.index:
    r_ID, UID, kcat, kcat_pred = df["reaction_id"][ind], df["uniprot_ac"][ind], df["value"][ind], df["pred_value"][ind]
    help_df = data_test.loc[data_test["Reaction ID"] == r_ID].loc[data_test["y_true"] == np.round(kcat,5)]
    for ind2 in help_df.index:
        data_test["y_pred_ENKIE"][ind2] = kcat_pred
data_test.loc[~pd.isnull(data_test["y_pred_ENKIE"])]
```

```python
y_true = np.array(data_test["y_true"].loc[~pd.isnull(data_test["y_pred_ENKIE"])])
y_pred = np.array(data_test["y_pred"].loc[~pd.isnull(data_test["y_pred_ENKIE"])])
y_pred_enkie =np.array(data_test["y_pred_ENKIE"].loc[~pd.isnull(data_test["y_pred_ENKIE"])])
```

```python
r2_score(y_true, y_pred), r2_score(y_true, y_pred_enkie)
```
